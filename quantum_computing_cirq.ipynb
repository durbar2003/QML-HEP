```python
import cirq
import numpy as np

# Define the qubits
qubits = [cirq.LineQubit(i) for i in range(5)]

# Define the circuit
circuit = cirq.Circuit()

# Apply the Hadamard operation to every qubit
for qubit in qubits:
    circuit.append(cirq.H(qubit))

# Apply the CNOT operations
circuit.append(cirq.CNOT(qubits[0], qubits[1]))
circuit.append(cirq.CNOT(qubits[1], qubits[2]))
circuit.append(cirq.CNOT(qubits[2], qubits[3]))
circuit.append(cirq.CNOT(qubits[3], qubits[4]))

# Apply the SWAP operation
circuit.append(cirq.SWAP(qubits[0], qubits[4]))

# Apply the X-rotation operation
circuit.append(cirq.rx(0.5 * np.pi).on(qubits[2]))

# Print the circuit
print(circuit)

# Draw the circuit
from cirq.contrib.jupyter import CircuitDiagramInfo
diagram = CircuitDiagramInfo(circuit)
diagram.generate_svg()
```
